## Nanda Land Cover Data

The data is really segmented across different proportions, so we are going to combine like datasets for easier analysis. Combine all the forest proportions into one forest, keep development into their own columns, combine [unclassified, open water, perennial snow/ice, barren land, and wetlands], combine hay/pasture, cultivated crops into one.

This will result in a dataset that has census tracts, years, and then these land cover variables (all proportions):
- Unclassified and harsh living conditions
- Open urban space
- Low intensity developed urban space
- Medium intensity developed urban space
- High intensity developed urban space
- Forested areas
- Cultivation and agrigultural areas

```python
# First limit the nanda for the variables I want (all proportion-based values)
nanda_lc_2016 = nanda_lc_2016[['tract_fips', 'year_intp','prop_value_0', 'prop_value_11', 'prop_value_12', 'prop_value_21', 'prop_value_22', 'prop_value_23', 'prop_value_24', 
                               'prop_value_31', 'prop_value_41', 'prop_value_42', 'prop_value_43', 'prop_value_52', 
                               'prop_value_71', 'prop_value_81', 'prop_value_82', 'prop_value_90', 'prop_value_95']]
```

```python
nanda_lc_2016.info()
```

```python
# no nulls and all floating points (the variables interested in)
# Get unique values of year_intp, so I can figure out which years to filter down for (to match the ltdb data)
nanda_lc_2016['year_intp'].unique()
```

```python
#Okay so yearly, and we are going to only select the ones from 2001, 2012, and 2016
nanda_lc_2016 = nanda_lc_2016[nanda_lc_2016['year_intp'].isin({2001, 2012, 2016})]
```

```python
#Now to start combining the props into summed sets based on the grouping above
nanda_lc_2016['unclassified'] = nanda_lc_2016['prop_value_0'] + nanda_lc_2016['prop_value_11'] + nanda_lc_2016['prop_value_12'] + nanda_lc_2016['prop_value_31'] + nanda_lc_2016['prop_value_90'] + nanda_lc_2016['prop_value_95']
nanda_lc_2016['forested'] = nanda_lc_2016['prop_value_41'] + nanda_lc_2016['prop_value_42'] + nanda_lc_2016['prop_value_43'] + nanda_lc_2016['prop_value_52'] + nanda_lc_2016['prop_value_71']
nanda_lc_2016['cultivated'] = nanda_lc_2016['prop_value_81'] + nanda_lc_2016['prop_value_82']
```

```python
#Rename the other prop columns (and the tracts and year column)
nanda_lc_2016 = nanda_lc_2016.rename(columns = {'tract_fips': 'tractid_nanda', 'year_intp': 'year_nanda', 
                                               'prop_value_21': 'open_urban_space', 'prop_value_22': 'low_development', 'prop_value_23': 'medium_development', 'prop_value_24': 'high_development'})
```

```python
nanda_lc_2016 = nanda_lc_2016[['tractid_nanda', 'year_nanda', 'open_urban_space', 
                              'low_development', 'medium_development', 'high_development', 'unclassified', 'forested', 'cultivated' ]]
```

```python
#just checking to make sure the proporions make sense
for i in nanda_lc_2016:
    print(i, max(nanda_lc_2016[i]), min(nanda_lc_2016[i]) )
```

```python
#Everything looks good, now to merge this dataset with all the census tracts data, as well as the specific cities
#I am going to fudge the numbers a little bit in this dataset, to match it with the year column in the ltdb. That way there are two joins available on the big dataset, and the correct tract is with the correct tract
nanda_lc_2016['year_nanda'] = nanda_lc_2016['year_nanda'].replace(2001, 2000)
nanda_lc_2016['year_nanda'] = nanda_lc_2016['year_nanda'].replace(2016, 2019)
```

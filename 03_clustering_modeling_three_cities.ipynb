### What are we looking for

We will be searching for clusters of potential abject poverty indicators in the socioeconomic, census-based variables. These will be values with very low standardized scores (i.e. centroids of clusters below the mean, with the exception of population, unclear what that one needs to be). We'll only be focusing on those clusters. The EDA portion will analyze the differences between cities in the data in these clusters, and then how the land cover proportion of census tracts is different in these clusters between these cities.

```python
##potential values to pull out to tune the model more - 'pop', 'incpc', 'hinc', 'phs', 'mrent', 'mhmval', 'pown', 'pmulti'
X_sea = seattle_df.drop(columns = ['Unnamed: 0', 'tractid', 'tract', 'county', 'state', 'year', 'GEOID10', 'NAMELSAD10', 'tractid_nanda', 'year_nanda'])
X_nyc = nyc_df.drop(columns = ['Unnamed: 0', 'tractid', 'tract', 'county', 'state', 'year', 'GEOID10', 'NAMELSAD10', 'tractid_nanda', 'year_nanda'])
X_dc = dc_df.drop(columns = ['Unnamed: 0', 'tractid', 'tract', 'county', 'state', 'year', 'GEOID10', 'NAMELSAD10', 'tractid_nanda', 'year_nanda'])
```

```python
# Use that "gridsearch" way of pulling best clusters out to identify the best number of clusters to fit
#Seattle first
scores = []

for k in range(2, 12):
    cl = KMeans(n_clusters=k, random_state=42)
    cl.fit(X_sea)
    inertia = cl.inertia_
    sil = silhouette_score(X_sea, cl.labels_)
    
    scores.append([k, inertia, sil])
    
score_df = pd.DataFrame(scores)
score_df.columns = ['k', 'inertia', 'silhouette_score']
```

```python
score_df # Eight clusters seems to fit Seattle the best, though it's not fantastic (not bad though, just meh)
```

```python
km = KMeans(n_clusters=8, random_state=42)
km.fit(X_sea)
seattle_df['cluster'] = km.predict(X_sea)
```

```python
seattle_df.head()
```

```python
seattle_df.drop(columns = ['Unnamed: 0', 'tractid', 'tract', 'county', 'state', 'year', 'GEOID10', 'NAMELSAD10', 'tractid_nanda', 'year_nanda']).groupby('cluster').mean()
```

Very quick way to find poverty clusters is to see the high values of ppov. There are two clusters that really take the cake for it, cluster 0 and cluster 4. So let's look at these ones to pick out the differences between them.

Cluster 0 is low population, average per capita income, low household income, lower than average high school grads, low rent value, average home values, very low home ownership, very high multi-home living. The census tracts they belong in are very very high developed areas, signifying a city, with very low forested and agricultural land.

Cluster 4 is average population, very low income per capita, household income, but has a high number of high school graduates, rent and home values are low, and not a lot of people own homes. Higher than average people live in multi-family dwellings, but not as many as cluster 0. Noteworthy is that they are mainly living in medium development areas, so likely lower-class suburbs.

I wanted to point out Cluster 7, which are probably more rural poor tracts. They don't have as many people living in poverty, but they do have low levels of income, and high levels of home ownership, so it is a different kind of poverty than people living closer into the MSA.

## Exploratory Data Analysis Using Markov Chains and Spatial Analysis

**To Do Here**

- Make Markov Chain transition. This is essentially a table that shows the proportion of clusters that occupy each space in that snapshot, and how that changes through time. An example is if in the year 2000, Seattle's Cluster 7 accounts for 50 percent of the total census tract in the MSA, and then by 2019 Cluster 7 accounts for 40 percent, we could determine that the evolution of the MSA landscape in Seattle is going away from the rural poor areas.
- Spatial EDA. Plot the clusters' locations on a map, to figure out contiguity, identify geographic clusters, and establish if there is a spatial relationship between clusters, to help profile the full MSA as well as pockets of Cluster profiles within the MSA. 

```python
#Adam's part: Markov Chain Matrix
#It's not a true Markov Chain model, but instead we're coopting the use of the matrix to create a table of transitions
#This will be the plot of part of the whole over each snapshot in time

#Bring in the just exported datasets, the three csvs of US cities
seattle_clusters = pd.read_csv("./data/seattle_clusters.csv")
nyc_clusters = pd.read_csv("./data/nyc_clusters.csv")
dc_clusters = pd.read_csv("./data/dc_clusters.csv")
```

```python
#Get the value counts (normalized) for each csv's snapshot in time
#create markov dataframe for each city
seattle_mk_df = pd.DataFrame()
nyc_mk_df = pd.DataFrame()
dc_mk_df = pd.DataFrame()
```

```python
# Organize the new table by proportion of cluster

for i in seattle_clusters['year'].unique():
    seattle_mk_df[i] = seattle_clusters[seattle_clusters['year'] == i]['cluster'].value_counts(normalize=True)
    
for i in nyc_clusters['year'].unique():
    nyc_mk_df[i] = nyc_clusters[nyc_clusters['year'] == i]['cluster'].value_counts(normalize=True)

for i in dc_clusters['year'].unique():
    dc_mk_df[i] = dc_clusters[dc_clusters['year'] == i]['cluster'].value_counts(normalize=True)
```

```python
seattle_mk_df
```

These are the Markov Chains ^^ and let's look at the changes in them. The only noteworthy thing is that Cluster 3 seems to be decreasing over time, though it had plateaued in 2019. Cluster 3 seems to be more forested census tracts, so it seemingly has lost land to development in other clusters. Cluster 1, which is mainly low density urban space, seemed to take this, signifying some early development of more heavily forested census tracts.

```python
nyc_mk_df
```

Cluster 7, the impoverished urban inner core, seems to have marginally increased over time. It could mean that more urban areas have transitioned to this type of profile, or the areas in the NY MSA just improved across the board and dropped more census tracts into this category in later years.

```python
dc_mk_df
```

DC has a few more noteworthy transitions. Cluster 4, the suburban less-developed areas, has increased, likely as that urban sprawl has enveloped more census tracts in the MSA. In addition, Cluster 3, the profile of sparsely populated forested and agricultural areas has decreased, signifying transitions into more urbanized spaces in these former lands. Finally Cluster 1 has increased and Cluster 2 has decreased. The dilineation between these urbanized poor areas is how developed the census tracts are. Cluster 1 increasing signifies the MSA having more numerous urban inner core poor areas, and less numerous populated outer core areas.
